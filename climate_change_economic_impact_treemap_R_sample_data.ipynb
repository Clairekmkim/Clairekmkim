```r
# Install and load required packages
if (!require(ggplot2)) install.packages("ggplot2")
if (!require(treemapify)) install.packages("treemapify")
if (!require(showtext)) install.packages("showtext")
if (!require(sysfonts)) install.packages("sysfonts")
library(ggplot2)
library(treemapify)
library(showtext)
library(sysfonts)

# Load Noto Sans KR Bold font
font_add_google("Noto Sans KR", "Noto Sans KR Bold", db_cache = FALSE)
showtext_auto()

# Data creation
climate_data <- data.frame(
  category = c(rep("Damage Costs", 5), rep("Response Costs", 4), rep("Economic Benefits", 2)),
  subcategory = c("Natural Disaster Recovery", "Agricultural Damage", "Increased Medical Costs", "Sea Level Rise Response", "Insurance Losses",
                  "Renewable Energy Transition", "Carbon Capture Technology", "Electric Vehicle Adoption", "International Agreement Implementation",
                  "Energy Cost Savings", "Job Creation"),
  value = c(125, 30, 23, 40, 135,
            75, 5, 7, 10,
            100, 50)
)

# Define color palette with more intense red
color_palette <- c("Damage Costs" = "#FF0000", "Response Costs" = "#4A90E2", "Economic Benefits" = "#66C172")

# Create the treemap with bolder font
treemap_plot <- ggplot(climate_data, aes(area = value, fill = category, label = subcategory)) +
  geom_treemap(layout = "squarified", alpha = 0.9, colour = "white", size = 3) +
  geom_treemap_text(layout = "squarified",
                    colour = "white",
                    place = "centre",
                    size = 10,
                    grow = TRUE,
                    min.size = 4,
                    reflow = TRUE,
                    family = "Noto Sans KR Bold",
                    fontface = "bold") +
  scale_fill_manual(values = color_palette) +
  labs(title = "Climate Change Related Costs and Benefits") +
  theme(legend.position = "bottom",
        plot.title = element_text(hjust = 0.5, size = 18, face = "bold", family = "Noto Sans KR Bold"),
        plot.background = element_rect(fill = "white", colour = NA),
        panel.background = element_rect(fill = "white", colour = NA),
        text = element_text(family = "Noto Sans KR Bold", face = "bold"),
        legend.text = element_text(face = "bold"),
        legend.title = element_text(face = "bold")) +
  guides(fill = guide_legend(title = "Category"))

# Display the plot
print(treemap_plot)

# Save the plot with higher resolution (optional)
# ggsave("rounded_treemap_bold.png", treemap_plot, width = 12, height = 10, dpi = 300)
# print("Treemap has been saved as 'rounded_treemap_bold.png' file.")
```
